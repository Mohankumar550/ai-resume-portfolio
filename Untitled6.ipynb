```python
import torch
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    TrainingArguments,
    Trainer,
    BitsAndBytesConfig
)
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
import os

print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"CUDA device: {torch.cuda.get_device_name()}")
```

```python
import os

model_id = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
save_dir = "./models/TinyLlama-1.1B-Chat-v1.0"

# Create save directory
os.makedirs(save_dir, exist_ok=True)

# Load tokenizer
print("Loading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(model_id)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# Configure quantization for GPU
if torch.cuda.is_available():
    print("Loading model with 4-bit quantization...")
    bnb_config = BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_quant_type="nf4",
        bnb_4bit_compute_dtype=torch.float16,
        bnb_4bit_use_double_quant=True,
    )

    model = AutoModelForCausalLM.from_pretrained(
        model_id,
        quantization_config=bnb_config,
        device_map="auto",
        trust_remote_code=True
    )
    model = prepare_model_for_kbit_training(model)
else:
    print("Loading model on CPU...")
    model = AutoModelForCausalLM.from_pretrained(
        model_id,
        torch_dtype=torch.float32,
        device_map="cpu",
        trust_remote_code=True
    )

print("Model loaded successfully!")
```

```python
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj"],
    lora_dropout=0.1,
    bias="none",
    task_type="CAUSAL_LM"
)

# Apply LoRA
model = get_peft_model(model, lora_config)
model.print_trainable_parameters()
```

```python
def format_prompt(instruction, output):
    return f"""### Instruction:
{instruction}

### Response:
{output}<|endoftext|>"""

# Process the data
processed_data = []
for item in data:
    text = format_prompt(item["instruction"], item["output"])
    processed_data.append({"text": text})

# Create dataset
dataset = Dataset.from_list(processed_data)
print(f"Processed dataset size: {len(dataset)}")
print("Sample processed text:")
print(dataset[0]["text"][:200] + "...")
```

```python
def tokenize_function(examples):
    # Tokenize the text
    tokenized = tokenizer(
        examples["text"],
        truncation=True,
        padding=False,
        max_length=512,
        add_special_tokens=True,
    )

    # For causal language modeling, labels are the same as input_ids
    tokenized["labels"] = tokenized["input_ids"].copy()

    return tokenized

# Tokenize the dataset
print("Tokenizing dataset...")
tokenized_dataset = dataset.map(
    tokenize_function,
    batched=True,
    remove_columns=dataset.column_names,
    desc="Tokenizing"
)

print(f"Tokenized dataset: {tokenized_dataset}")
```

```python
from transformers import DataCollatorForLanguageModeling

# Data collator
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False,  # We're doing causal LM, not masked LM
)
```

```python
training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=1,
    gradient_accumulation_steps=4,
    num_train_epochs=5,
    learning_rate=1e-4,
    fp16=torch.cuda.is_available(),
    logging_steps=1,
    save_steps=50,
    eval_strategy="no",
    save_strategy="epoch",
    warmup_steps=2,
    report_to="none",
    remove_unused_columns=False,
    max_grad_norm=1.0,            # Add gradient clipping
    weight_decay=0.01,            # Reduce overfitting


)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator,
)
```

```python
print("Starting training...")
print("This may take 5-15 minutes depending on your hardware...")

# Train the model
trainer.train()

print("Training completed!")
```

```python
print("Testing the fine-tuned model...")



def generate_response(prompt, max_length=100):
    # Format the prompt
    formatted_prompt = f"### Instruction:\n{prompt}\n\n### Response:\n"

    # Tokenize
    inputs = tokenizer(formatted_prompt, return_tensors="pt")

    # Move to device
    if torch.cuda.is_available():
        inputs = {k: v.to("cuda") for k, v in inputs.items()}

    # Generate
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_length=max_length,
            temperature=0.3,
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id,
            eos_token_id=tokenizer.eos_token_id,
        )

    # Decode
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract just the response part
    if "### Response:" in response:
        response = response.split("### Response:")[1].strip()

    return response

# Test with some prompts
test_prompts = [
    "Hello",
    "Who are you?",
    "Tell me about machine learning",
    "How are you doing today?"
]

print("\n" + "="*50)
print("TESTING FINE-TUNED MODEL")
print("="*50)

for prompt in test_prompts:
    print(f"\nPrompt: {prompt}")
    print(f"Response: {generate_response(prompt)}")
    print("-" * 30)

print("\nTesting complete!")
```

```python
def generate_precise_response(prompt, max_new_tokens=30):  # Limit length
    formatted_prompt = f"### Instruction:\n{prompt}\n\n### Response:\n"
    inputs = tokenizer(formatted_prompt, return_tensors="pt")

    if torch.cuda.is_available():
        inputs = {k: v.to("cuda") for k, v in inputs.items()}

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,  # Limit new tokens
            temperature=0.3,  # Lower temperature = more focused
            do_sample=True,
            top_p=0.9,  # Nucleus sampling for precision
            repetition_penalty=1.2,
            pad_token_id=tokenizer.eos_token_id,
            eos_token_id=tokenizer.eos_token_id,
            early_stopping=True
        )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    if "### Response:" in response:
        response = response.split("### Response:")[1].strip()

    # Stop at first sentence for precision
    if ". " in response:
        response = response.split(". ")[0] + "."

    return response
```

```python
import shutil
import os

def save_model_checkpoint(checkpoint_path, save_path):
    """Save a specific checkpoint to local directory"""

    # Create save directory if it doesn't exist
    os.makedirs(save_path, exist_ok=True)

    # Copy checkpoint files
    if os.path.exists(checkpoint_path):
        shutil.copytree(checkpoint_path, os.path.join(save_path, "checkpoint"), dirs_exist_ok=True)
        print(f"Checkpoint saved to {save_path}")
    else:
        print(f"Checkpoint path {checkpoint_path} not found")

# Save your latest checkpoint
save_model_checkpoint("./results/checkpoint-805", "./my_saved_model")
```
